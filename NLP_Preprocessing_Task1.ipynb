```python
import re
import string
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('punkt_tab')   # needed in newer environments
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text = "Hello!!! This is a sample NLP task :) Visit https://example.com NOW!!! #NLP #AI 123"

print("Original Text:\n", text)
```

```python
def to_lower(text):
    return text.lower()

text = to_lower(text)
print("Lowercase:\n", text)
```

```python
def remove_urls(text):
    return re.sub(r'http\S+|www\S+', '', text)

text = remove_urls(text)
print("Without URLs:\n", text)
```

```python
def remove_numbers(text):
    return re.sub(r'\d+', '', text)

text = remove_numbers(text)
print("Without Numbers:\n", text)
```

```python
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

text = remove_punctuation(text)
print("Without Punctuation:\n", text)
```

```python
def remove_extra_spaces(text):
    return " ".join(text.split())

text = remove_extra_spaces(text)
print("Cleaned Spaces:\n", text)
```

```python
tokens = word_tokenize(text)
print("Tokens:\n", tokens)
```

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(tokens):
    return [word for word in tokens if word not in stop_words]

tokens = remove_stopwords(tokens)
print("Without Stopwords:\n", tokens)
```

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_words(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]

tokens = lemmatize_words(tokens)
print("After Lemmatization:\n", tokens)
```

```python
final_text = " ".join(tokens)
print("Final Cleaned Text:\n", final_text)
```
